```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data from csv
data = pd.read_csv('Bank_Personal_Loan_Modelling.csv')
data 
```

```python
# clean the data
data = data.dropna()
data = data.drop_duplicates()
```

```python
#  Statistical Calculations: Mean, Median, Std Dev, Skewness, and Kurtosis
def calculate_statistics(data):
    """
    Calculate and display key statistics (mean, median, std, skewness, kurtosis) for numeric columns.
    
    Parameters:
    data: DataFrame with numeric columns to analyze.
    """
    stats_df = pd.DataFrame({
        'Mean': data.mean(),
        'Median': data.median(),
        'Std Dev': data.std(),
        'Skewness': data.skew(),
        'Kurtosis': data.kurtosis()
    })
    return stats_df
calculate_statistics(data)
```

```python
# basic statastics using describe function 
stat = data.describe()
stat
```

```python
# infprmation about data
info = data.info()
info
```

```python
# correlation matrix 
corr = data.corr()
corr
```

```python
# 1. Relational Graph (Scatter Plot): Income vs. Average Credit Card Spending
def plot_income_vs_ccavg(data):
    """
    Plot a scatter plot showing the relationship between Income and CCAvg (Average Credit Card Spending).
    
    Parameters:
    - data: DataFrame containing 'Income' and 'CCAvg' columns.
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(data['Income'], data['CCAvg'], alpha=0.6)
    plt.title('Income vs. Average Credit Card Spending (CCAvg)')
    plt.xlabel('Income')
    plt.ylabel('CCAvg (Average Credit Card Spending)')
    plt.show()

plot_income_vs_ccavg(data)
```

```python
# 2. Categorical Graph (Bar Chart): Education vs. Personal Loan Acceptance
def plot_education_vs_loan(data):
    """
    Plot a bar chart comparing the acceptance of personal loans across education levels.
    
    Parameters:
    - data: DataFrame containing 'Education' and 'Personal Loan' columns.
    """
    loan_by_education = data.groupby('Education')['Personal Loan'].value_counts().unstack()
    loan_by_education.plot(kind='bar', stacked=True, figsize=(8, 6))
    plt.title('Personal Loan Acceptance by Education Level')
    plt.xlabel('Education Level')
    plt.ylabel('Number of Individuals')
    plt.legend(title='Personal Loan')
    plt.show()

plot_education_vs_loan(data)
```

```python
# 3. Focused Statistical Graph (Heatmap): Correlation Matrix for Selected Columns
def plot_focused_correlation_heatmap(correlation_matrix):
    """
    Plot a heatmap of the selected correlation matrix to visualize meaningful relationships between key variables.
    
    Parameters:
    - correlation_matrix: Correlation matrix DataFrame with selected columns to visualize.
    """
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Focused Correlation Matrix Heatmap')
    plt.show()

# Selecting meaningful columns for the heatmap
selected_columns = ['Income', 'CCAvg', 'Personal Loan', 'Education', 'Mortgage', 'CD Account', 'CreditCard']
selected_correlation_matrix = data[selected_columns].corr()
plot_focused_correlation_heatmap(selected_correlation_matrix)
```
